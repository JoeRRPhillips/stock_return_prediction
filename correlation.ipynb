```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
# Assess correlations across the entire dataset
train_df = pd.read_csv("data/train_data.csv")
val_df = pd.read_csv("data/validation_data.csv")
df = pd.concat([train_df, val_df])

labels_colname = "direction_tp1"
features_df = df.copy()
labels_df = features_df.pop(labels_colname)
```

```python
correlation_df = features_df.corr()
columns = correlation_df.columns

correlation_df \
    .style.background_gradient(cmap="coolwarm", axis=None, vmin=-1, vmax=1) \
    .highlight_null(null_color='#f1f1f1') \
    .set_precision(2)
```
